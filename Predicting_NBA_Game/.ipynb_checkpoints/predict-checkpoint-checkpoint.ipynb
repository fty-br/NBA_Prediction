### Outliers

```python
colonnes=categorical_columns +['won']
df_num= df.drop(columns=colonnes)
```

```python
df_num
```

```python
variables = df_num.columns
percentage_outliers = []

for variable in variables:
    q1 = df_num[variable].quantile(0.25)
    q3 = df_num[variable].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df_num[(df_num[variable] < lower_bound) | (df_num[variable] > upper_bound)]

    percentage = (len(outliers) / len(df_num)) * 100
    percentage_outliers.append(percentage)

outliers_df = pd.DataFrame({
    'Variable': variables,
    'Pourcentage d\'Outliers': percentage_outliers
})

columns_per_plot = 20
num_plots = len(variables) // columns_per_plot + (1 if len(variables) % columns_per_plot != 0 else 0)

for i in range(num_plots):
    start_idx = i * columns_per_plot
    end_idx = start_idx + columns_per_plot

    plt.figure(figsize=(16, 8))
    sns.barplot(x='Variable', y='Pourcentage d\'Outliers', data=outliers_df.iloc[start_idx:end_idx], palette='coolwarm')
    plt.xlabel('Variable')
    plt.ylabel('Pourcentage d\'Outliers')
    plt.title(f'Pourcentage d\'Outliers par Variable (Partie {i + 1})')
    plt.ylim(0, 100)
    plt.xticks(rotation=90)

    for j in range(start_idx, end_idx):
        if j < len(percentage_outliers):
            plt.text(j - start_idx, percentage_outliers[j] + 2, f'{percentage_outliers[j]:.2f}%', ha='center', va='bottom', fontsize=9, rotation=90)

    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
df_rolling = df[list(selected_columns) + ["won", "team", "season"]]

def find_team_averages(team):
    # Conserver les colonnes 'season' et 'team'
    season_team_info = team[['team']].iloc[0]
    # Supprimer les colonnes 'won', 'season' et 'team'
    team = team.drop(columns=['team'])
    # Calculer la moyenne mobile sur 10 périodes
    rolling = team.rolling(10).mean()
    # Réintégrer les informations 'season' et 'team'
    for col in season_team_info.index:
        rolling[col] = season_team_info[col]
    return rolling
```

```python
df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)
```

```python
def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_col(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: shift_col(x, col_name))

df["home_next"] = add_col(df, "home")
df["team_opp_next"] = add_col(df, "team_opp")
df["date_next"] = add_col(df, "date")
```

### Compare all the models

```python
from sklearn.linear_model import RidgeClassifier, LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
```

```python

# Fonction pour créer un modèle de réseau de neurones
def create_nn_model():
    model = Sequential()
    model.add(Dense(64, input_dim=len(selected_columns), activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Changer l'activation pour correspondre à votre problème (classification binaire/multiclasse)
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Initialisation des modèles
models = {
    "RidgeClassifier": RidgeClassifier(),
    "LogisticRegression": LogisticRegression(max_iter=1000),  # Augmentation des itérations pour convergence
    "RandomForestClassifier": RandomForestClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier(),
    "AdaBoostClassifier": AdaBoostClassifier(),
    "SVC": SVC(),
    "KNeighborsClassifier": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "MLPClassifier": MLPClassifier(max_iter=1000),
    "XGBClassifier": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "LGBMClassifier": LGBMClassifier(),
    "CatBoostClassifier": CatBoostClassifier(learning_rate=0.05, l2_leaf_reg=5, iterations=200, depth=6, border_count=50),
    "NeuralNetwork": KerasClassifier(build_fn=create_nn_model, epochs=10, batch_size=10, verbose=0)
}

# Dictionnaire pour stocker les précisions
accuracies = {}

# Boucle pour backtester chaque modèle
for model_name, model in models.items():
    predictions = backtest(full, model, selected_columns)
    accuracy = accuracy_score(predictions["actual"], predictions["prediction"])  # Assurez-vous que 'actual' et 'prediction' sont corrects
    accuracies[model_name] = accuracy

# Afficher les précisions
for model_name, accuracy in accuracies.items():
    print(f"{model_name} Accuracy: {accuracy:.2f}")
```

```python
# Calculate the correlation matrix
correlation_matrix = df_num.corr()

# Create a mask to identify the features with a correlation coefficient greater than or equal to 0.75
high_correlation_mask = correlation_matrix >= 0.75

# Identify and list the highly correlated features
highly_correlated_features = []

for feature in high_correlation_mask.columns:
    correlated_with = high_correlation_mask.index[high_correlation_mask[feature]].tolist()
    for correlated_feature in correlated_with:
        if feature != correlated_feature and (correlated_feature, feature) not in highly_correlated_features:
            highly_correlated_features.append((feature, correlated_feature))

# Print the highly correlated features
print("Highly correlated features:")
for feature1, feature2 in highly_correlated_features:
    print(f"{feature1} and {feature2}")
```

```python
# Supprimer une des colonnes hautement corrélées
columns_to_drop = set()
for feature1, feature2 in highly_correlated_features:
    if feature1 not in columns_to_drop and feature2 not in columns_to_drop:
        columns_to_drop.add(feature2)

full_reduced= full.drop(columns=columns_to_drop)
```
